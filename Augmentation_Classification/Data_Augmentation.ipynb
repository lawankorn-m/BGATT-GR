# à¸ºUndersampling and Oversampling Augmentation

```python
from sklearn.datasets import load_iris
from sklearn.datasets import load_breast_cancer
from collections import Counter
import numpy as np
from imblearn.over_sampling import BorderlineSMOTE
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd

# Manually add label 0 or 1 to the csv input file
f = pd.read_csv('PCA230-BiGRU4000_Full_Training_Set.csv')
X_resampled = f.iloc[:, 1:-1].values
y_resampled = f.iloc[:, -1].values

number = 220 # Define the size

# Undersampling
rus = RandomUnderSampler(sampling_strategy={0: number}, random_state=42)
X_resampled, y_resampled = rus.fit_resample(X_resampled, y_resampled)

# Oversampling
borderline_smote = BorderlineSMOTE()
X_resampled, y_resampled = borderline_smote.fit_resample(X_resampled, y_resampled)
pd.concat([pd.DataFrame(X_resampled), pd.DataFrame(y_resampled)], axis=1).to_csv('R0.csv', header=True, index=True)
```
